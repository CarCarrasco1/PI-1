 **NOTA EXPLICATIVA DE LO SIGUIENTE:**

 
En esta línea de código, la función apply se utiliza para aplicar una función lambda a cada fila del DataFrame. La función lambda toma cada fila (row) y verifica si el valor de duration es NaN (pd.isnull(row['duration'])). Si el valor de duration es NaN, verifica si el valor correspondiente en bis es numérico (isinstance(row['bis'], (int, float))) y no es NaN (pd.notnull(row['bis'])). Si ambas condiciones se cumplen, se reemplaza el valor de duration con el valor correspondiente en bis. Si no se cumplen las condiciones, se mantiene el valor original de duration.

De esta manera, se crea una nueva columna duration que contiene los mismos valores que la columna original, pero con los valores NaN reemplazados por los valores correspondientes de bis donde sea posible.

```python
#Reemplazo los valores Nulos, de la columna 'duration' por los valores numericos que se encuentran en la misma posicion en la columna bis
#completando asi los valores cambiado de una a otra. 
df3['duration'] = df3.apply(lambda row: row['bis'] if pd.isnull(row['duration']) and pd.notnull(row['bis']) and isinstance(row['bis'], (int, float)) else row['duration'], axis=1)
df3.head()
```

```python
#Prosigo a borrar la columna bis
df3.drop('bis', axis=1, inplace=True)
```

**EXPLICACIÓN DEL SIGUIENTE CÓDIGO**


En esta línea de código, loc se utiliza para seleccionar las filas en las que la columna 'RATING' que es del tipo object y comienza con números (df['rating'].apply(lambda x: isinstance(x, str) and x[0].isdigit())). Luego, se selecciona la columna 'RATING' en esas filas ('rating'), y se les asigna el valor nulo np.nan.

Después de ejecutar esta línea de código, todos los valores de 'RATING' que son del tipo object y comienzan con números se reemplazarán por valores nulos en esas filas.

```python
df3.loc[df3['rating'].apply(lambda x: isinstance(x, str) and x[0].isdigit()), 'rating'] = np.nan
```

Renombro la columna de **'rating'** por **'score'** para evitar las confusiones en el futuro.

Realizo un promedio de los **'score'**

```python
#Realizo un promedio de la columna score, agrupada por promedio.
promedio_score = df_score.groupby('id')['score'].mean()
```

```python
#Redondeo a las segunda decimal.
promedio_final = promedio_score.round(2)
```

```python
#Realizo un dataframe pequeño con los promedios con los respectivos id
score_mean = pd.DataFrame({'id': promedio_final.index, 'score_mean': promedio_final.values})
```
